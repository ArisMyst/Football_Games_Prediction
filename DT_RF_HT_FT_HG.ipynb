```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("PL_20200307.csv")
df.head()
```

```python
inputs = df.drop('FTHG',axis='columns')
```

```python
target = df['FTHG']
```

```python
from sklearn.preprocessing import LabelEncoder
le_HOMETEAM = LabelEncoder()
le_AWAYTEAM = LabelEncoder()
le_HTR = LabelEncoder()
```

```python
inputs['HOMETEAM_n'] = le_HOMETEAM.fit_transform(inputs['HOMETEAM'])
inputs['AWAYTEAM_n'] = le_AWAYTEAM.fit_transform(inputs['AWAYTEAM'])
inputs['HTR_n'] = le_HTR.fit_transform(inputs['HTR'])
```

```python
inputs
```

```python
inputs_n = inputs.drop(['HOMETEAM','AWAYTEAM','HTR','FTR','FTAG','HC','AC','HY', 'AY', 'DIV'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(max_depth=6)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_n, target, test_size=0.2, random_state=0)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
